<h1 align="center">Diagnostico de Pneumonia usando RX</h1>
A pneumonia é uma condição inflamatória do pulmão afetando principalmente os pequenos sacos aéreos conhecidos como alvéolos. Tipicamente, os sintomas incluem alguma combinação de tosse seca ou produtiva, dor torácica, febre e dificuldade em respirar. A gravidade é variável.

A pneumonia geralmente é causada por infecção de vírus ou bactérias e, mais raramente por outros microrganismos, certos medicamentos e condições, como doenças autoimunes. Os fatores de risco incluem outras doenças pulmonares, como fibrose cística, DPOC e asma, diabetes, insuficiência cardíaca, história de tabagismo, falta de habilidade para tossir, como seguir um derrame ou um sistema imunológico fraco. O diagnóstico geralmente é baseado nos sintomas e exame físico. A radiografia de tórax, exames de sangue e cultura do escarro podem ajudar a confirmar o diagnóstico. A doença pode ser classificada por onde foi adquirida com pneumonia associada à comunidade, hospital ou à saúde.

<h2>Criação do modelo.</h2>



```python
image_height = 224 #150
image_width = 224 #150
batch_size = 32 #10
no_of_epochs  = 40 #10
```

```python
model = Sequential()
model.add(Conv2D(32,(3,3),input_shape=(image_height,image_width,3),activation='relu'))
model.add(Conv2D(32,(3,3),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.2))
model.add(Conv2D(64,(3,3),activation='relu'))
model.add(Conv2D(64,(3,3),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.2))
model.add(Conv2D(128,(3,3),activation='relu'))
model.add(Conv2D(128,(3,3),activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.2))
model.add(Flatten())
model.add(Dense(units=128,activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(units=1,activation='sigmoid'))
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

**Vamos proceder a Data Augmentation antes de fornecer as imagens ao modelo.**

*Será aplicado Rescaling,Shearing,zooming and Rotation nos dados de Treino e só rescaling nos dados de teste*

```python
train_datagen = ImageDataGenerator(rescale=1./255,
                                   rotation_range=15,
                                   shear_range=0.2,
                                   zoom_range=0.2
                                   )

test_datagen = ImageDataGenerator(rescale=1./255)

val_datagen = ImageDataGenerator(rescale=1./255)
```

```python
training_set = train_datagen.flow_from_directory(PATH+'/train',
                                                 target_size=(image_width, image_height),
                                                 batch_size=batch_size,
                                                 class_mode='binary')  #tb podia ser class_mode='categorical'

test_set = test_datagen.flow_from_directory(PATH+'/test',
                                            target_size=(image_width, image_height),
                                            batch_size=batch_size,
                                            class_mode='binary')

# não implementado --->
#val_set = test_datagen.flow_from_directory(PATH+'/val',
#                                            target_size=(image_width, image_height),
#                                            batch_size=1,
#                                            shuffle=False,
#                                            class_mode='binary')
```

```python
reduce_learning_rate = ReduceLROnPlateau(monitor='loss',
                                         factor=0.1,
                                         patience=2,
                                         cooldown=2,
                                         min_lr=0.00001,
                                         verbose=1)

callbacks = [reduce_learning_rate]
```

<h2>Treino do Modelo.</h2>

```python
history = model.fit_generator(training_set,
                    steps_per_epoch=5235//batch_size,
                    epochs=no_of_epochs,
                    validation_data=test_set,
                    validation_steps=624//batch_size,
                    callbacks=callbacks
                   )
```

```python

print(test_set.class_indices)
```

```python
acc = history.history['acc']
val_acc = history.history['val_acc']
loss = history.history['loss']
val_loss = history.history['val_loss']

epochs = range(1, len(acc) + 1)
fig = plt.figure(figsize=(16,9))

plt.subplot(1, 2, 1)
plt.plot(epochs, loss, 'bo', label='Training loss')
plt.plot(epochs, val_loss, 'b', label='Validation loss')
plt.title('Training and validation loss')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(epochs, acc, 'bo', label='Training acc')
plt.plot(epochs, val_acc, 'b', label='Validation acc')
plt.title('Training and validation accuracy')
plt.legend()
plt.show()
```

```python
predictions = model.predict_generator(test_set, steps=16, verbose=1)
```
