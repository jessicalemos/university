```python
import os

from sklearn.ensemble import BaggingRegressor
from sklearn.svm import SVC
from sklearn.ensemble import BaggingClassifier
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.svm import SVR
from sklearn import linear_model
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from sklearn import preprocessing
import pandas as pd
from sklearn import datasets, linear_model
from sklearn import svm,model_selection
from sklearn.model_selection import train_test_split, KFold, cross_val_score, cross_val_predict, GridSearchCV
from sklearn.preprocessing import Binarizer
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
from sklearn.naive_bayes import GaussianNB
from sklearn.feature_selection import VarianceThreshold,SelectFromModel
from sklearn.naive_bayes import GaussianNB
from sklearn.feature_selection import SelectKBest, f_classif, chi2
from sklearn.svm import SVC

data = pd.read_csv("train_data.csv",sep=",")
test= pd.read_csv("test_data.csv",sep=",")
test_target= pd.read_csv("sample_submission.csv",sep=",")

data['Work load Average/day '] = [x.replace(',', '') for x in data['Work load Average/day ']]
data['Work load Average/day '] = data['Work load Average/day '].astype(int)

target=data['Absent']
data=data.drop(['Absent'],1)
test_target=test_target['Absent']
data=data.drop(['ID'],1)
test=test.drop(['ID'],1)

#selectkbest
selector = SelectKBest(f_classif, k=5)
selector.fit(data, target)
#features selection
cols = selector.get_support(indices=True)
cols_names = list(data.columns[cols])
data = data[cols_names]
test = test[cols_names]

def modelos(data):
	model_RFP = RandomForestRegressor(n_estimators = 500, random_state = 1).fit(data,target)
	predictions_RFP = model_RFP.predict(test)
	df_new_rfp_pred = pd.DataFrame({'actual':test_target,'predicted':predictions_RFP})
	result = accuracy_score(df_new_rfp_pred['actual'],round(df_new_rfp_pred['predicted']))
	print("Random Forest %0.6f" % result)

	model_DT = KNeighborsClassifier().fit(data,target)
	predictions_DT = model_DT.predict(test)
	df_new_dt_pred_l = pd.DataFrame({'actual':test_target,'predicted':predictions_DT})
	result = accuracy_score(df_new_dt_pred_l['actual'],round(df_new_dt_pred_l['predicted']))
	print("Nearest Neighbors %0.6f" % result)

	model_DT = svm.SVC(gamma="auto").fit(data,target)
	predictions_DT = model_DT.predict(test)
	df_new_dt_pred_l = pd.DataFrame({'actual':test_target,'predicted':predictions_DT})
	result =accuracy_score(df_new_dt_pred_l['actual'],round(df_new_dt_pred_l['predicted']))
	print("SVM %0.6f" % result)

	model_DT = LogisticRegression(solver='lbfgs',max_iter=5000).fit(data,target)
	predictions_DT = model_DT.predict(test)
	df_new_dt_pred_l1 = pd.DataFrame({'actual':test_target,'predicted':predictions_DT})
	result =accuracy_score(df_new_dt_pred_l1['actual'],round(df_new_dt_pred_l1['predicted']))
	print("LogisticRegression %0.6f" % result)
    
	model_DT = DecisionTreeRegressor(random_state = 1).fit(data,target)
	predictions_DT = model_DT.predict(test)
	df_new_dt_pred_l = pd.DataFrame({'actual':test_target,'predicted':predictions_DT})
	result = accuracy_score(df_new_dt_pred_l['actual'],round(df_new_dt_pred_l['predicted']))
	print("Decision Tree %0.6f" % result)    

	model_DT = AdaBoostClassifier().fit(data,target)
	predictions_DT = model_DT.predict(test)
	df_new_dt_pred_l = pd.DataFrame({'actual':test_target,'predicted':predictions_DT})
	result =accuracy_score(df_new_dt_pred_l['actual'],round(df_new_dt_pred_l['predicted']))
	print("AdaBoost %0.6f" % result)   

	model_DT = GaussianNB().fit(data,target)
	predictions_DT = model_DT.predict(test)
	df_new_dt_pred_l = pd.DataFrame({'actual':test_target,'predicted':predictions_DT})
	result =accuracy_score(df_new_dt_pred_l['actual'],round(df_new_dt_pred_l['predicted']))
	print("Naive Bayes %0.6f" % result)

	model_DT = linear_model.BayesianRidge().fit(data,target)
	predictions_DT = model_DT.predict(test)
	df_new_dt_pred_l = pd.DataFrame({'actual':test_target,'predicted':predictions_DT})
	result =accuracy_score(df_new_dt_pred_l['actual'],round(df_new_dt_pred_l['predicted']))
	print("Bayesian Ridge %0.6f" % result)  
   
	model_DT = BaggingClassifier(base_estimator=SVC(gamma='auto'),n_estimators=10, random_state=0).fit(data,target)
	predictions_DT = model_DT.predict(test)
	df_new_dt_pred_l = pd.DataFrame({'actual':test_target,'predicted':predictions_DT})
	result =accuracy_score(df_new_dt_pred_l['actual'],round(df_new_dt_pred_l['predicted']))
	print("Bagging SVC %0.6f" % result)  
    
	model_DT = BaggingRegressor(base_estimator=DecisionTreeRegressor(random_state = 1),n_estimators=10, random_state=0).fit(data,target)
	predictions_DT = model_DT.predict(test)
	df_new_dt_pred_l = pd.DataFrame({'actual':test_target,'predicted':predictions_DT})
	result =accuracy_score(df_new_dt_pred_l['actual'],round(df_new_dt_pred_l['predicted']))
	print("Bagging DecisionTree %0.6f" % result)  
    
       
    
testN=test
dataN = data
##normalization
normalizer = preprocessing.Normalizer(norm='l2') 
values_normalized = normalizer.transform(dataN.values)
dataN = pd.DataFrame(values_normalized, columns=dataN.columns)
values_normalized = normalizer.transform(test.values)
test = pd.DataFrame(values_normalized, columns=test.columns)
print("---------Normalization---------")
modelos(dataN)

test=testN

##standardization
scaler = preprocessing.MinMaxScaler(feature_range=(0, 1))
values_standardized = scaler.fit_transform(data)
data = pd.DataFrame(values_standardized, columns=data.columns);
values_standardized = scaler.fit_transform(test)
test = pd.DataFrame(values_standardized, columns=test.columns);
print("--------Standardization--------")
modelos(data)
```
